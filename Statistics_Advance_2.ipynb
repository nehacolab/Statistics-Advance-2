```python
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    :param sample: List or array of sample data
    :param population_mean: Known population mean
    :param population_std: Known population standard deviation
    :param alpha: Significance level (default is 0.05)
    :return: Z-score, p-value, and test conclusion
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpret results
    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference."
    else:
        result = "Fail to reject the null hypothesis: No significant difference found."

    return z_score, p_value, result

# Example usage
sample_data = [55, 60, 65, 62, 58, 59, 61, 63, 57, 64]
pop_mean = 60
pop_std = 5

z_score, p_value, conclusion = z_test(sample_data, pop_mean, pop_std)
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

def z_test(sample, population_mean, population_std, alpha=0.05):
    """Perform a one-sample two-tailed Z-test."""
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return z_score, p_value, result

def independent_t_test(sample1, sample2, alpha=0.05):
    """Perform an independent two-sample T-test."""
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return t_stat, p_value, result

def paired_t_test(before, after, alpha=0.05):
    """Perform a paired sample T-test."""
    t_stat, p_value = stats.ttest_rel(before, after)

    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return t_stat, p_value, result

def f_test(sample1, sample2, alpha=0.05):
    """Perform an F-test to compare the variances of two samples and visualize the results."""
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1

    p_value = 2 * (1 - stats.f.cdf(f_stat, dfn, dfd))
    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    plt.figure(figsize=(8, 5))
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, dfn, dfd)
    plt.plot(x, y, label='F-Distribution')
    plt.axvline(f_stat, color='r', linestyle='--', label=f'F-statistic: {f_stat:.4f}')
    plt.title("F-Test: Comparing Variances")
    plt.xlabel("F-Statistic")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    return f_stat, p_value, result

def chi_square_variance_test(sample, population_variance, alpha=0.05):
    """Perform a hypothesis test for population variance using a Chi-square distribution."""
    sample_size = len(sample)
    sample_variance = np.var(sample, ddof=1)

    chi_square_stat = (sample_size - 1) * sample_variance / population_variance

    lower_critical = stats.chi2.ppf(alpha / 2, df=sample_size - 1)
    upper_critical = stats.chi2.ppf(1 - alpha / 2, df=sample_size - 1)

    result = "Reject the null hypothesis" if chi_square_stat < lower_critical or chi_square_stat > upper_critical else "Fail to reject the null hypothesis"
    return chi_square_stat, lower_critical, upper_critical, result

def z_test_proportions(success1, size1, success2, size2, alpha=0.05):
    """Perform a Z-test for comparing proportions between two groups."""
    prop1 = success1 / size1
    prop2 = success2 / size2

    pooled_prop = (success1 + success2) / (size1 + size2)
    standard_error = np.sqrt(pooled_prop * (1 - pooled_prop) * (1/size1 + 1/size2))

    z_score = (prop1 - prop2) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return z_score, p_value, result

def simulate_and_test():
    """Simulate random data from a normal distribution and perform hypothesis testing."""
    np.random.seed(42)
    sample_data1 = np.random.normal(loc=50, scale=10, size=30)
    sample_data2 = np.random.normal(loc=55, scale=12, size=30)

    print("\n--- Independent T-Test for Mean Comparison ---")
    t_stat, p_value, conclusion = independent_t_test(sample_data1, sample_data2)
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}, Conclusion: {conclusion}")

    print("\n--- Variance Test using Chi-Square Distribution ---")
    chi_stat, lower_crit, upper_crit, chi_conclusion = chi_square_variance_test(sample_data1, population_variance=100)
    print(f"Chi-Square Statistic: {chi_stat:.4f}, Lower Critical Value: {lower_crit:.4f}, Upper Critical Value: {upper_crit:.4f}, Conclusion: {chi_conclusion}")

    print("\n--- Z-Test for Proportions ---")
    z_stat, p_val, z_conclusion = z_test_proportions(success1=30, size1=100, success2=40, size2=120)
    print(f"Z-Statistic: {z_stat:.4f}, P-value: {p_val:.4f}, Conclusion: {z_conclusion}")

    print("\n--- F-Test for Comparing Variances ---")
    f_stat, f_p_val, f_conclusion = f_test(sample_data1, sample_data2)
    print(f"F-Statistic: {f_stat:.4f}, P-value: {f_p_val:.4f}, Conclusion: {f_conclusion}")

    print("\n--- Visualizing Data Distribution ---")
    plt.figure(figsize=(8, 5))
    plt.hist(sample_data1, alpha=0.6, bins=15, label='Sample 1')
    plt.hist(sample_data2, alpha=0.6, bins=15, label='Sample 2')
    plt.legend()
    plt.title("Histogram of Simulated Data")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()

# Run the simulation and tests
simulate_and_test()
```
